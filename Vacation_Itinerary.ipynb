```python
# Create DataFrame called hotel_df to store hotel names along with city, country, max temp, and coordinates.
hotel_df = city_data_df[["City", "Country", "Max Temp","Current Description", "Lat", "Lng","Rain inches (last 3 hrs)","Snow inches (last 3 hrs)"]].copy()
hotel_df["Hotel Name"] = ""
hotel_df.head(10)
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Set parameters to search for a hotel.
params = {
    "radius": 5000,
    "type": "lodging",
    "key": g_key
}
```

```python
# Iterate through the DataFrame.
for index, row in hotel_df.iterrows():
    
    # Grab the first hotel from the results and store the name.
    try:
        # Get the latitude and longitude.
        lat = row["Lat"]
        lng = row["Lng"]

        # Add the latitude and longitude to the params dictionary as values to
        # the location key.
        params["location"] = f"{lat},{lng}"

        # Use the search term: "lodging" and our latitude and longitude.
        base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
        # Make request and get the JSON data from the search.
        hotels = requests.get(base_url, params=params).json()
        # Grab the first hotel from the results and store the name.
        hotel_df.loc[index, "Hotel Name"] = hotels["results"][0]["name"]
    
    except (IndexError):
        print("Hotel not found... skipping.")
```

```python
hotel_df.head()
```

```python
vacation_stop_1=hotel_df.loc[hotel_df['City']=='Saint-Jean-De-Maurienne']
vacation_stop_2=hotel_df.loc[hotel_df['City']=='Saint-Joseph']
vacation_stop_3=hotel_df.loc[hotel_df['City']=='Nago']
vacation_stop_4=hotel_df.loc[hotel_df['City']=='Roma']
```

```python
itinerary_df= pd.concat([vacation_stop_1,vacation_stop_2,vacation_stop_3,vacation_stop_4])
itinerary_df.reset_index(drop=True)
```

```python
info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Current Weather </dt><dd>{Current Description} and {Max Temp} °F</dd>
</dl>
"""
```

```python
# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in itinerary_df.iterrows()]
```

```python
# Add a heatmap of temperature for the vacation spots and a pop-up marker for each city.
locations = itinerary_df[["Lat", "Lng"]]
max_temp = itinerary_df["Max Temp"]
fig = gmaps.figure(center=(30.0, 31.0), zoom_level=1.5)
itinerario_itfr= gmaps.directions_layer(sjd_maurienne,roma,show_markers=False,waypoints=[s_joseph,nago],travel_mode= 'DRIVING')
marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(itinerario_itfr)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
